```python
import requests
import numpy as np
import pandas as pd
```

```python
laguardia_data2 = pd.read_csv("Laguardia_data.csv")
cp_data2 = pd.read_csv("Central_park_data2.csv")
combined = pd.concat([laguardia_data2, cp_data2]).reset_index(drop = True)

combined.head(30)

# drop STATION, REPORT_TYPE, SOURCE because it is not important
combined = combined.drop(columns = ["STATION"])
combined
```

```python
# use pd.timestamp to get year, month, and day

combined["DATE"] = combined.DATE.map(pd.Timestamp)
combined["YEAR"] = combined.DATE.map(lambda dt: dt.year)   #extract year from DATE
combined["MONTH"] = combined.DATE.map(lambda dm: dm.month)
combined["DAY"] = combined.DATE.map(lambda dd: dd.day)
combined = combined.sort_values(by = ["YEAR", "MONTH", "DAY"], ascending = False)
combined = combined.reset_index(drop = True)
combined.tail(20)
 
combined
```

```python
# find columns with NAs
check_missing = combined.isna().any()
missing = check_missing[check_missing == True]
missing = list(missing.index)
missing
```

```python
# drop columns that are not important

combined_col = list(combined.columns)

combined = combined.drop(columns = ["DATE", "REPORT_TYPE", "SOURCE"])

remove_list = []
for i in range(len(combined_col)):
    if "Backup" in combined_col[i]:
        remove_list.append(combined_col[i])
        
remove_list

combined = combined.drop(columns = remove_list)

combined
```

```python
# get remaining NAs

check_missing = combined.isna().any()
missing = check_missing[check_missing == True]
missing = list(missing.index)
missing
```

```python
# use isna().sum() to determine which columns is more than 50% NAs, then remove it
remove_list = []

for col in combined.columns:
    if combined[col].isna().sum()> (len(combined)/2):
        remove_list.append(col)
        
remove_list
    
```

```python
combined = combined.drop(columns = remove_list)
combined.columns
```

```python
check_missing = combined.isna().any()
missing = check_missing[check_missing == True]
missing
combined.dtypes
```

```python
# looking at original data, remove columns not needed

combined = combined.drop(columns = ["HourlyWindDirection","HourlySkyConditions", "REM", "REPORT_TYPE.1", "SOURCE.1", "HourlyAltimeterSetting"])
combined.head(20)
```

```python
# convert object columns to float so that mean can be calculated

columns = ["HourlyDewPointTemperature", 
           "HourlyDryBulbTemperature", 
           "HourlyPrecipitation", 
           "HourlyRelativeHumidity", 
           "HourlySeaLevelPressure",
           "HourlyStationPressure", 
           "HourlyVisibility",
           "HourlyWetBulbTemperature", 
           "HourlyWindSpeed"]

# see percentage of data missing
for i in range(len(columns)):
    col = columns[i]
    print(col, ":", combined[col].isna().sum())
    print("Percentage of data missing:", (combined[col].isna().sum()/len(combined))*100)

value_list = []                         # unable to convert columns to float, need to look into values
for i in range(len(columns)):
    col = columns[i]
    value = combined[col].value_counts()
    value_list.append(value.index.unique())
print(value_list)
```

```python
value_list = []
for i in range(len(columns)):
    col = columns[i]
    combined[col] = combined[col].map(lambda x: str(x))  # column seems to contain different data types, resulting in a lot of NAs when converted to float
    combined[col] = combined[col].map(lambda x: x.rstrip("s"))  # suspect values - can be treated the same as actual value for now
    combined[col] = combined[col].map(lambda x: str(0) if x == "T" else x)     # T indicates trace amounts - can be treated as 0s
    combined[col] = combined[col].map(lambda x: x.rstrip("V"))     # no mention of what it means in documentation, assume is the same as suspect values
    combined[col] = combined[col].map(lambda x: str(0) if x == "*" or x == "0.020.01" else x)    # Amounts included in following measurement; time distribution unknown - treat as zero
    combined[col] = combined[col].map(lambda x: float(x))
    
```

```python
# see percentage of data missing again to see if there are any changes
combined.head(20)

for i in range(len(columns)):
    col = columns[i]
    print(col, ":", combined[col].isna().sum())
    print("Percentage of data missing:", (combined[col].isna().sum()/len(combined))*100)
    
combined.dtypes
```

```python
# eliminate NAs with by getting mean by month

avg = combined.groupby(["YEAR","MONTH", "DAY"]).mean() # Can create dataframe for daily data set
avg.reset_index(inplace=True)
avg.columns = ["YEAR", "MONTH", "DAY", "DailyDewPointTemperature(celsius)", "DailyDryBulbTemperature(celsius)", "DailyPrecipitation(cm)(Hourlymean)", "DailyRelativeHumidity", "DailySeaLevelPressure(inches)", "DailyStationPressure(inches)", "DailyVisibility", "DailyWetBulbTemperature(celsius)", "DailyWindSpeed(miles/hr)"]
avg.tail(70)  # precipitation in inches to hundredths, need to convert to inches by multiplying by 100 and then cm
```

```python
# temperature is in Fahrenheit, convert to Celsius

tempcol = ["DailyDewPointTemperature(celsius)", "DailyDryBulbTemperature(celsius)", "DailyWetBulbTemperature(celsius)"]

for i in range(len(tempcol)):
    col = tempcol[i]
    avg[col] = avg[col].map(lambda x:(x - 32)* 5/9)
    
avg["DailyPrecipitation(cm)(Hourlymean)"] = avg["DailyPrecipitation(cm)(Hourlymean)"].map(lambda x: (x*100)*2.54)
av
```

```python
cplaguardia_data = pd.read_csv("https://www.ncei.noaa.gov/access/services/data/v1?dataset=daily-summaries&stations=USW00014732,USW00094728&startDate=2011-05-02&endDate=2021-04-19&dataTypes=AWND,PRCP,SNOW,SNWD,TMAX,TMIN,WDF2,WDF5,WSF2,WSF5&format=csv&units=metric")
cplaguardia_data.columns = ["STATION", "DATE", "AWND(metres/sec)", "PRCP(cm)", "SNOW(cm)", "SNWD(cm)", "TMAX(celsius)", "TMIN(celsius)", "WDF2(degrees)", "WDF5(degrees)", "WSF2(metres/sec)", "WSF5(metres/sec)"]
cplaguardia_data = cplaguardia_data.drop(columns = ["STATION"])
cplaguardia_data   #PRCP, SNOW, and SNWD is actually in mm, convert to cm later
```

```python
# create YEAR, MONTH, DAY columns

cplaguardia_data["DATE"] = cplaguardia_data.DATE.map(pd.Timestamp)
cplaguardia_data["YEAR"] = cplaguardia_data.DATE.map(lambda dt: dt.year)   #extract year from DATE
cplaguardia_data["MONTH"] = cplaguardia_data.DATE.map(lambda dm: dm.month)
cplaguardia_data["DAY"] = cplaguardia_data.DATE.map(lambda dd: dd.day)
cplaguardia_data = cplaguardia_data.sort_values(by = ["YEAR", "MONTH", "DAY"], ascending = True)
cplaguardia_data = cplaguardia_data.reset_index(drop = True)
cplaguardia_data.head(20)

cplaguardia_data = cplaguardia_data.drop(columns = ["DATE"])
 
cplaguardia_data
```

```python
# check for dtype and NAs in columns

print(cplaguardia_data.dtypes)
```

```python
# remove duplicates in data first (two observations for one day due to getting data from two stations)

cpl_avg = cplaguardia_data.groupby(["YEAR","MONTH", "DAY"]).mean() # Can create dataframe for daily data set
cpl_avg.reset_index(inplace=True)

# change columns using mm to cm

mm_col = ["PRCP(cm)", "SNOW(cm)", "SNWD(cm)"]
for i in range(3):
    col = mm_col[i]
    cpl_avg[col] = cpl_avg[col].map(lambda x: x/10)
    
cpl_avg
```

```python
for col in cpl_avg.columns:
    print(col, ":", cpl_avg[col].isna().sum())
    print("Percentage of data missing:", (cpl_avg[col].isna().sum()/len(cpl_avg))*100, "\n")
```

```python
# combine both datasets to get final dataset
# then create csv in case data is lost

merged_df = pd.merge(cpl_avg, avg, how='left', on=["YEAR", "MONTH","DAY"])
merged_df
merged_df.to_csv('finalsproject_dataset.csv')
```
